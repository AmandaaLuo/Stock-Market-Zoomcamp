```python
#Q4 52-weeks range ratio (2023) for the selected stocks
data = yf.download(['2222.SR','BRK-B','AAPL', 'MSFT','GOOG','JPM'], start='2023-01-01', end = '2024-01-01')
```

```python
data.tail()
```

```python
# Initialize an empty dictionary to store the range for each ticker
ranges = {}
for ticker in data['Adj Close'].columns:
    # Calculate the max and min values for the ticker
    max_value = data['Adj Close'][ticker].max()
    min_value = data['Adj Close'][ticker].min()
    # Calculate the range for the ticker
    if max_value != 0:  # Avoid division by zero
        range_value = (max_value - min_value) / max_value
    else:
        range_value = None
    # Store the range in the dictionary
    ranges[ticker] = range_value
```

```python
ranges
```

```python
max_range_ticker = max(ranges, key=ranges.get)
max_range_ticker
```

```python
max_range_value = ranges[max_range_ticker]
max_range_value.round(2)
```

```python
#Q5 Dividend Yield
tickers = ['2222.SR', 'BRK-B', 'AAPL', 'MSFT', 'GOOG', 'JPM']
dividends = {}
closing_prices = {}
dividend_yield = {}

for ticker in tickers:
    ticker_dividends = yf.Ticker(ticker).dividends
    ticker_dividends_2023 = ticker_dividends.loc['2023']
    dividends[ticker] = ticker_dividends_2023
    total_dividends = ticker_dividends.loc['2023'].sum()

    data_2023 = yf.download(ticker, start='2023-01-01', end='2023-12-31')
    last_day_closing_price = data_2023['Adj Close'].iloc[-1]


    dividend_yield[ticker] = total_dividends / last_day_closing_price

```

```python
dividends
```

```python
dividend_yield
```

```python
max_yield_ticker = max(dividend_yield, key=dividend_yield.get)

max_yield_value = dividend_yield[max_yield_ticker]
```
