```python
def AveragePricecalculation (price_range):
      if '-' in price_range:  # If a range is given
        prices = price_range.split('-')
        prices = [price.strip().replace('$', '').replace(',', '') for price in prices if price.strip()]
        if len(prices) == 2:
            return np.mean([float(price) for price in prices])
        else:
            return np.nan
      elif price_range.strip() != '':  # If only one price is provided
        return float(price_range.strip().replace('$', '').replace(',', ''))
      else:  # If no price is specified
        return np.nan

ipos_2023['Avg_price'] = ipos_2023['Price Range'].apply(AveragePricecalculation)
```

```python
# revise ticker list
tickers = list(stacked_ipos_df.Symbol)
tickers.append('HOVR')
```

```python
# Initialize a DataFrame to store the min_dates and growth data for all tickers
all_data_df = pd.DataFrame()

for ticker in tickers:
    print(ticker)
    ticker_df = yf.download(tickers=ticker, period='max', interval='1d')

    # Store the min_date for the current ticker
    min_date = ticker_df.index.min()

    # Generate additional columns for growth_future_1d to growth_future_30d
    for days in range(1, 31):
        ticker_df[f'growth_future_{days}d'] = ticker_df['Adj Close'].shift(-days) / ticker_df['Adj Close']

    # Add the ticker and min_date columns to the ticker_df
    ticker_df['Ticker'] = ticker
    ticker_df['Min_Date'] = min_date

    # Append the ticker_df to the all_data_df
    all_data_df = pd.concat([all_data_df, ticker_df])
```

```python
# Filter the data to include only the rows on or after the min_date for each ticker
all_data_df = all_data_df[all_data_df.index >= all_data_df['Min_Date']]

growth_percentiles = all_data_df[[f'growth_future_{days}d' for days in range(1, 31)]].describe()
```

```python
# Define a function to calculate CCI for each stock
def calculate_CCI(data, window=20):
    # Calculate Typical Price (TP)
    TP = (data['High'] + data['Low'] + data['Close']) / 3
    # Calculate Simple Moving Average (SMA) of TP
    SMA = TP.rolling(window=window).mean()
    # Calculate Mean Deviation (MD) from SMA
    MD = TP.rolling(window=window).apply(lambda x: pd.Series(x - x.mean()).abs().mean(), raw=True)
    # Calculate Commodity Channel Index (CCI)
    CCI = (TP - SMA) / (0.015 * MD)
    return CCI
```

```python
# Calculate CCI for each stock
data['CCI'] = data.groupby('Ticker')['Ticker'].transform(lambda x: calculate_CCI(x))

```

```python
# Define function to calculate expected gross profit
def calculate_gross_profit(data, threshold=200, initial_investment=1000):
    gross_profit = 0
    for date, group in data.groupby('Date'):
        # Filter Fridays
        if date.dayofweek != 4:
            continue

        for idx, row in group.iterrows():
            # Check if CCI > threshold
            if row['CCI'] > threshold:
                # Calculate profit
                profit = row['Adj Close'] - row['Open']
                gross_profit += profit * initial_investment / row['Adj Close']
                break  # Only invest once per Friday
    return gross_profit

# Calculate expected gross profit over many operations in 10 years
expected_gross_profit = calculate_gross_profit(data)
# Convert to thousands and round to the closest integer
expected_gross_profit_thousands = round(expected_gross_profit / 1000)

print("Expected gross profit (in thousands $):", expected_gross_profit_thousands)
```
